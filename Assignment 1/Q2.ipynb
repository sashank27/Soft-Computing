```python
import numpy
import matplotlib.pyplot as plt
```

```python
# Input data
x = numpy.array([75, 80, 93, 65, 87, 71, 98, 68, 84, 77])
y = numpy.array([82, 78, 86, 72, 91, 80, 95, 72, 89, 74])
```

```python
plt.scatter(x, y, color = "m", marker = "o", s = 30)
```

### Fitting a Least Square Line

```python
# Regression of Y on X
m = 10
a = 0
b = 0
c = 0
d = 0
for xi in x:
    a = a + xi
    c = c + xi*xi

for yi in y:
    b = b + yi
    
for i in range(0, len(x)):
    d = d + x[i]*y[i]
    
w1 = (a*b - d*m)/(a*a - c*m)
w0 = (a*d - b*c)/(a*a - c*m)

h = w0 + w1*x
plt.scatter(x, y, color = "m", marker = "o", s = 30)
plt.plot(x, h, color = "g")
plt.xlabel("Marks in SOC")
plt.ylabel("Marks in HUR")
plt.show()
```

```python
print(w0)
print(w1)
```

```python
# Regression of X on Y
m = 10
a = 0
b = 0
c = 0
d = 0
for yi in y:
    a = a + yi
    c = c + yi*yi

for xi in x:
    b = b + xi
    
for i in range(0, len(x)):
    d = d + x[i]*y[i]
    
w11 = (a*b - d*m)/(a*a - c*m)
w00 = (a*d - b*c)/(a*a - c*m)

h1 = w00 + w11*y
plt.scatter(x, y, color = "m", marker = "o", s = 30)
plt.plot(h1, y, color = "g")
plt.xlabel("Marks in SOC")
plt.ylabel("Marks in HUR")
plt.show()
```

```python
print(w00)
print(w11)
```

```python
plt.scatter(x, y, color = "m", marker = "o", s = 30)
plt.plot(x, h, color = "g")
plt.plot(h1, y, color = "b")
plt.grid(color="r")
plt.xlabel("Marks in SOC")
plt.ylabel("Marks in HUR")
plt.show()
```
