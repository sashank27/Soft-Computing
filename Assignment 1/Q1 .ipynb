```python
import numpy
import matplotlib.pyplot as plt
```

```python
# Input data
x = numpy.array([2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017])
y = numpy.array([61.2, 58.3, 67.1, 69.2, 68.9, 93.5, 89.1, 80.0, 92.3, 93.0, 97.0])
```

```python
plt.scatter(x, y, color = "m", marker = "o", s = 30)
```

```python
m = 11
a = 0
b = 0
c = 0
d = 0
for xi in x:
    a = a + xi
    c = c + xi*xi

for yi in y:
    b = b + yi
    
for i in range(0, len(x)):
    d = d + x[i]*y[i]
    
```

```python
w1 = (a*b - d*m)/(a*a - c*m)
w0 = (a*d - b*c)/(a*a - c*m)
print(w0)
print(w1)
```

```python
h = w0 + w1*x
plt.scatter(x, y, color = "m", marker = "o", s = 30)
plt.plot(x, h, color = "g")
plt.grid(color="r")
plt.xlabel('Year')
plt.ylabel('Revenue in Billion')
plt.show()
```

```python
# Expected Revenue in 2019
print(w0 + w1 * 2019)
```
