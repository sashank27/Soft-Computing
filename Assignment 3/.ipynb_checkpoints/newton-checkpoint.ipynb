```python
import numpy as np
```

```python
def loss(x,y,theta):
    m,n = np.shape(x)
    cost_list = []
    for i in range(0,n): 
        x_0 = x[:,i].reshape((m,1))
        print(x_0)
        predicted = np.dot(x_0,theta[i])
        error = predicted - y
        cost = np.sum(error ** 2) /  m
        cost_list.append(cost)

    cost_list = np.array(cost_list).reshape((2,1))
    print(cost_list)
    return cost_list

def NewtonMethod(x,y,theta,maxIterations):
    m,n = np.shape(x)
    xTrans  = x.transpose()
    H       = 2 * np.dot(xTrans,x) / m
    Hinv    = np.linalg.inv(H)
    thetaPrev = np.zeros_like(theta)
    best_iter = maxIterations
    for i in range(0,maxIterations):
        cost = loss(x,y,theta)
        theta  = theta - np.dot(Hinv,cost)

    return theta
```

```python
x = np.array([[-1.7,-1.5],[-1.0,-0.3],[1.7,1.5],[-1.2,-0.7],[0.6,0.1]])  
y = np.array([0.3 ,0.07,-0.2,0.07,0.03 ])

#Normalising Inputs(2D input)
def normalise(inp):
    return np.array((inp-inp.mean())/inp.std())
x = normalise(x)
y = y*1.0

print(x.shape)
y = y.reshape(y.shape[0],1)
print(y.shape)
theta = np.zeros(2)
theta = theta.reshape(theta.shape[0],1)
print(theta.shape)
NewtonMethod(x,y,theta,100)
```
